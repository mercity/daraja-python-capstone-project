```python
#this plot shows State in US with the highest population of women
plt.figure(figsize =(8,5))
sns.barplot(x="State", y="Population", data = breast_cancer_df)
plt.xticks(rotation = 90)
plt.title('The Population of Women in each state in US in 2011-2015')
```

```python
#this plot shows the State in US with the highest number of women that died of breast cancer
plt.figure(figsize =(8,6))
sns.barplot(x="State", y="DeathCount", data = breast_cancer_df)
plt.xticks(rotation = 90)
plt.title('The State with the highest number of Women that died of Breast Cancer in US in 2011-2015')
```

```python
#total number of population and women that survived breast cancer rate in US in the year 2011-2015
#total number of women who got affected with breast cancer and total number of women that died of breast cancer in the year 2011-2015 in US
breast_cancer_df.sum(axis = 0, skipna = True)
```

```python
#The number of people that got affected by breast cancer divided by the population of each state so that it can be used to get the state with the highest brest cancer rate
Rate = breast_cancer_df['Incidence']/breast_cancer_df['Population']
Rate
```

```python
#I added Rate as a new column to the dataframe
breast_cancer_df['Rate'] = Rate
breast_cancer_df
```

```python
#This shows the state with the highest breast cancer rate
plt.figure(figsize =(15,5))
sns.barplot(x="State", y="Rate", data = breast_cancer_df)
plt.title('Breast Cancer Rate')
plt.xticks(rotation = 90)
```

```python
#How many women survived breast cancer in us
plt.figure(figsize =(15,5))
sns.barplot(x="State", y="Alive", data = breast_cancer_df)
plt.title('The State that have the highest survival rate')
plt.xticks(rotation = 90)
```
