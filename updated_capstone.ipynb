```python
#These are the four countries with the most highest number of women population in US, i want to know which of the country have the highest breast cancer rate
data = {'Country': ['California', 'Utah', 'North Carolin', 'Florida'],
       'Population': [96473279, 66782188, 50613521, 50178974],
       'Incidence': [128452.0, 7239.0, 39269.0, 77149.0],
       'DeathCount': [22033, 13837, 12987, 14200],
       'Alive': [12541526.27, 721184.5, 4043286.08, 7526846.1]}
breast_cancer_df = pd.DataFrame(data, columns = ['Country', 'Population', 'Incidence','DeathCount','Alive'])
breast_cancer_df
```

```python
#The country with the highest rate of occurence in breast cancer
sns.scatterplot(x="Country",y="Incidence", data=breast_cancer_df)
plt.title('California have the highest breast cancer rate')
```

```python
#How many women survived breast cancer in us
sns.catplot(x="Country", y="Alive", data = breast_cancer_df, kind="bar", hue="Country")
plt.title('California has the highest number of women that survived breast cancer')
```
